```python
files = glob.glob("./playlists/*/*/*.csv")
id_track  = defaultdict(list) 
d_id  = defaultdict(set) 
```

```python
for file in files:
    df = pd.read_csv(file)
    df = df.drop(["Unnamed: 0","index_x"] ,1)
    
    for s in range(len(df)):
        id_track[df["id"][s]].append(df["track"][s])
    
    for s in range(len(df)-1):
        d_id[df["id"][s]].add(df["id"][s+1])

    for s in range(1,len(df)):
        d_id[df["id"][s]].add(df["id"][s-1])
```

```python
def get_DiGraph(d):
    G = nx.DiGraph()
    for k,v in d.items():
        G.add_node(k)
        for w in v:
            G.add_edge(k,w)
    return G
```

```python
G_id = get_DiGraph(d_id)
```

```python
print("使用曲数: %d" %G_id.number_of_nodes()) 
print("曲の繋がり数: %d" %G_id.number_of_edges()) 
```

```python
#エッジ数でノードをソートした上位10つ
top10= sorted( dict(G_id.degree).items(),key = lambda x:x[1],reverse=True)[:10]
```

```python
for i in range(len(top10)):
    print("%d位:%s, エッジ数:%s" %(i+1, id_track[top10[i][0]][0], top10[i][1]))
```

```python
Match = pd.DataFrame(list(G_id.edges()),columns = ['row','col'])
Match['exist_flg'] = 1
#pivot
tb = Match.pivot_table(index='row',columns='col',values='exist_flg',fill_value=0)
```

```python
tb.to_csv('outputs/Match.csv', encoding="shift-jis")
```

```python
# ノードの名前を変更する
mapping_label =[]
for node_name in G_id.nodes:
    mapping_label.append(id_track[node_name][0])
    
mapping_label = [w.replace('$', '') for w in mapping_label]
mapping = {k: v for k, v in zip(G_id.nodes, mapping_label)}
```

```python
(pd.DataFrame.from_dict(data=mapping, orient='index')
   .to_csv('outputs/id_track.csv', header=True, encoding="utf_8_sig"))
```
